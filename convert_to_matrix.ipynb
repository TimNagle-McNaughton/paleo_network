```python
# general imports
import pandas as pd
```

```python
path_to_csv = r'E:/Git_repos/paleo_network/test.csv'
```

```python
# load csv to dataframe, ignore header rows
df = pd.read_csv(path_to_csv)
```

```python
# fill NaN values with 0
df = df.fillna(0)

# convert any floats to integers
df.iloc[:,1:] = df.iloc[:,1:].astype(int)
```

```python
# save to a textfile matrix
df.to_csv(r'E:/Git_repos/paleo_network/test.txt', sep=' ', index=False)
```
